```python
import numpy.matlib as np
import matplotlib.pyplot as plt

class Mountain_car:
    def __init__(self):
        V=np.linspace(-0.07,0.07,100)
        self.V=np.around(V,3)
        P=np.linspace(-1.2,0.5,100)
        self.P=np.around(P,2)
        self.A=[-1,0,1]
        self.m=np.zeros((100,100))
        self.policy=np.zeros((100,100))
    def reward(self,curr_p):
        if curr_p==0.5:
            return 100
        else:
            return -1
    def get_vindex(self,vel):
        error=1
        
        for i in range(100):
            if(self.V[i]==vel):
                return i
        for i in range(100):
            if(abs(self.V[i]-vel)<error):
                error=abs(self.V[i]-vel)
                error_index=i
        return error_index
    def get_pindex(self,pos):
        error=1
        
        for i in range(100):
            if(self.P[i]==pos):
                return i
        for i in range(100):
            if(abs(self.P[i]-pos)<error):
                error=abs(self.P[i]-pos)
                error_index=i
        return error_index
    def next_statevelocity(self,curr_v,acc,curr_p):
        v=curr_v+(0.001*acc)-0.0025*np.cos(3*curr_p)
        v=np.around(v,3)
        if(v<-0.07):
            return -0.07
        elif(v>0.07):
            return 0.07
        else:
            return v
    def next_stateposition(self,curr_v,curr_p):
        x=curr_v+curr_p
        x=np.around(x,2)
        if(x<-1.2):
            return -1.2
        else:
            return x
    def get_action(self,curr_p,curr_v):
        next_v=[]
        pos=self.next_stateposition(curr_v,curr_p)
        k=self.get_pindex(pos)
        for i in range(3):
            vel=self.next_statevelocity(curr_v,i-1,curr_p)
            j=self.get_vindex(vel)
            next_v.append(self.m[k,j])
        return [max(next_v),np.argmax(next_v)-1,pos,k]
    def drive(self):
        p=-0.5
        v=0
        x=self.get_pindex(p)
        y=self.get_vindex(v)
        position=[]
        velocity=[]
        i=0
        while(p!=0.5):
            position.append(p)
            velocity.append(v)
            Act=self.get_action(p,v)
            #print(Act)
            Imm_r=self.reward(p)
            #print(Imm_r)
            p_new=Act[2]
            v_new=self.next_statevelocity(v,Act[1],p)
            x_new=Act[3]
            y_new=self.get_vindex(v_new)
            self.m[x,y]=Imm_r+(0.9*Act[0])
            self.policy[x,y]=Act[1]
            plt.plot(velocity,position,'ro')
            #plt.show()
            #print(self.m)
            p=p_new
            v=v_new
            x=x_new
            y=y_new
            #print(p,v)
        ve=[]
        po=[]
        i=0
        while(i<1000):
            ve.append(velocity[i])
            po.append(position[i])
            i=i+1
        plt.plot(po,ve,'ro')
        plt.show()
        #1print('hurrray')
        #np.reshape(self.policy,-1)
        with open("optimal_value.txt","wb") as f:
            for line in self.m:
                np.savetxt(f, line, fmt='%.2f')
        with open("optimal_policy.txt","wb") as f:
            for line in self.policy:
                np.savetxt(f, line, fmt='%.2f')
        #print(self.m)
        #print(self.policy)
o=Mountain_car()
o.drive()
```
